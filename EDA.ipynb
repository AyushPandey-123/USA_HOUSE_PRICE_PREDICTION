## EDA

### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

### Importing all the processed Data into a Dataframe

```python
dir_path = "Processed Data"
path = []
for i in os.listdir(dir_path):
    path.append(os.path.join(dir_path,i))
```

```python
df_full = []

for i in path:
    print(i)
    df_full.append(pd.read_csv(i))
```

```python
df_final = pd.concat(df_full,ignore_index=False).groupby('DATE').sum()
```

### Desne Plots for the input variables

```python
plt.figure(figsize=(10,6))
for col in df_final.columns.to_list():
    plt.figure(figsize=(4, 4))
    sns.displot(data=df_final, x=col, kde =True)
    plt.xlabel(f"{col}")
    plt.ylabel("Count")
    plt.title(f'Plot for {col}')
    plt.show()
```

### Box Plot

```python
plt.figure(figsize=(10,6))
for col in df_final.columns.to_list():
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df_final, x=col)
    plt.xlabel(f"{col}")
    plt.title(f'Plot for {col}')
    plt.show()
```

### Outliers are present in 
###### 1. Brick Price
###### 2. Employment Ratio
###### 3. Monthly Housing Supply
###### 4. New Houses Under Construction
###### 5. Personal Saving Rate
###### 6. Unemployment Rate

### Removing Outlier

#### Brick Price

```python
df_final['Brick_Price'].skew()
```

###### The data is right skewed

```python
df_final['Brick_Price'] = np.log(df_final['Brick_Price'])
```

```python
df_final['Brick_Price'].hist()
```

```python
plt.boxplot(df_final['Brick_Price'],vert=False)
plt.xlabel('Brick_Price')
plt.title('Box Plot of Brick_Price')
plt.show()
```

##### Employment Ratio

```python
df_final['Em_Ratio'].skew()
```

###### The data is left skewed

```python
Q1 = df_final["Em_Ratio"].quantile(0.25)
Q3 = df_final["Em_Ratio"].quantile(0.75)

IQR =Q3-Q1
print(f"First Quartile {Q1}")
print(f"Third Quartile {Q3}")
print(f"Inter Quartile Rate {IQR}")
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound {lower_bound}")
print(f"Upper bound {upper_bound}")
```

```python
df_final['Em_Ratio'] = df_final['Em_Ratio'].clip(lower=lower_bound)
```

```python
plt.boxplot(df_final['Em_Ratio'], vert=False)
plt.xlabel('Em_Ratio')
plt.title('Box Plot of emratio')
plt.show()
```
